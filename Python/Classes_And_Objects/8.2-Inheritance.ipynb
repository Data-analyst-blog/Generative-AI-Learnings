## Inheritance In Python

Inheritance is a fundamental concept in Object-Oriented Programming (OOPS) that allows a class to inherit attributes and methods from another class. This lesson covers single inheritance and multiple inheritance, demonstrating how to create and use them in Python.

```python
## Inheritance
## Parent Class
class Car:
    ## Initializer / Instance Attributes/Constructor
    def __init__(self, windows, doors, enginetype):
        self.windows= windows
        self.doors= doors
        self.enginetype= enginetype

    ## Instance Method
    def drive(self):
        print(f"The person will drive the {self.enginetype} car with {self.doors} doors and {self.windows} windows.")

```

```python
car1 = Car(4,5,"Petrol")
car1.drive()
```

```python
## Single Inheritance
class Tesla(Car):
    def __init__(self, windows, doors, enginetype, is_selfdriving):
        ## Calling the constructor of the parent class
        super().__init__(windows, doors, enginetype)
        self.is_selfdriving = is_selfdriving

        ## Creating a method for the child class
    def selfdriving(self):
        print(f"Tesla supports self driving: {self.is_selfdriving}")
       
```

```python
tesla1 = Tesla(4,5,"electric", True)
tesla1.selfdriving()
```

```python
tesla1.drive()
```

```python
## Multiple Inheritance
## When a class inherits from more than one base calss.
## Base Class 1
class Animal:

    ## Constructor/Initializing methods
    def __init__(self, name):
        self.name = name

    ## Instance methods
    def speak(self):
        print("Subclass must implement this method")


# Base Class 2
class Pet:
    def __init__(self, owner):
        self.owner = owner

## Derived / Child Class
class Dog(Animal, Pet):
    def __init__(self, name, owner):
        ## Calling the constructors of the base classes
        Animal.__init__(self, name)
        Pet.__init__(self, owner)

    def speak(self):
        return f"{self.name} says Woof! Woof!"

## Create an object

dog1 = Dog('Buddy','Yash')
print(dog1.speak())
print(f"Owner is {dog1.owner}")


```
