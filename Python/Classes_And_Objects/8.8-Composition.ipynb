#### COMPOSITION (HAS-A RELATIONSHIP)

A class uses another class

```python
class Engine:
    def start(self):
        print("Engine Started!")

class Car:
    def __init__(self):
        self.engine = Engine()
    def drive(self):
        self.engine.start()
        print("Car is moving!")

## Defining object
my_car = Car()
print(my_car.drive())
```
