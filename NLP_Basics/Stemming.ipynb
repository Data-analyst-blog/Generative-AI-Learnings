#### Stemming

**Stemming** is the process of reducing words to their root form by removing suffixes.
Example: running, runner, ran → run

```python
words = ["eating","eats","eaten","Writings","written","writing", "program", "programs", "programmer", "programming","history","historical","historian", "finally","finalize","finalized", "congratulate","congratulation","congratulations"]

## Lets findout the stems of the words using PorterStemmer
from nltk.stem import PorterStemmer
ps = PorterStemmer()

for word in words:
    print(f"{word} --> {ps.stem(word)}")
```

#### RegexpStemmer Class

The **RegexpStemmer** class in NLTK uses regular expression rules to strip word endings and reduce words to their stem.

Example: with pattern ing$, running → runn, playing → play

```python
from nltk.stem import RegexpStemmer
rs = RegexpStemmer('ing$|s$|e$|al$|ian$|ly$|ate$|ulate$')

for word in words:
    print(f"{word} --> {rs.stem(word)}")
```

#### Snowball stemmer

The **SnowballStemmer** is an improved, language-specific stemming algorithm in NLTK that produces more accurate stems than basic stemmers.

Example: running, runner, ran → run

```python
from nltk.stem import SnowballStemmer

ss = SnowballStemmer('english')

for word in words:
    print(f"{word} --> {ss.stem(word)}")
```
