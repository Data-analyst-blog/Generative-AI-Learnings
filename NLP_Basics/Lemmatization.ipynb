### Lemmatization

**Lemmatization** is the process of reducing a word to its meaningful base form (lemma) using vocabulary and grammar rules.
It produces real words, unlike stemming.

Example: running → run, better → good

```python
import nltk
nltk.download('wordnet')
nltk.download('omw-1.4')
from nltk.stem import WordNetLemmatizer


wnl = WordNetLemmatizer()
words = ["eating","eats","eaten","Writings","written","writing", "program", "programs", "programmer", "programming","history","historical","historian", "finally","finalize","finalized", "congratulate","congratulation","congratulations"]
for word in words:
    print(f"{word} --> {wnl.lemmatize(word, pos = 'v')}")
```
