### Named Entity Recognition

**Named Entity Recognition** (NER) is a natural language processing technique that automatically identifies and categorizes key information in text into predefined labels like names of people, organizations, locations, and dates.

Example: In the sentence "Elon Musk visited Berlin in 2024," an NER system would tag "Elon Musk" as a Person, "Berlin" as a Location, and "2024" as a Date.

```python
sentence ="""The Eiffel Tower is located in Paris. It is one of the most famous landmarks in the world. 
Many tourists visit the Eiffel Tower every year to admire its architecture and enjoy the stunning views of the city from its observation decks. 
It was built in 1889 and has since become a symbol of France and a must-see attraction for visitors.
"""
```

```python
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
nltk.download('maxent_ne_chunker')
nltk.download('words')
nltk.download('maxent_ne_chunker_tab')
import nltk
from nltk.tokenize import word_tokenize, sent_tokenize
```

```python
words = word_tokenize(sentence)
pos_tags = nltk.pos_tag(words)
named_entities = nltk.ne_chunk(pos_tags)
print(named_entities)
```

```python
named_entities.draw()  # This will open a window displaying the named entity tree visualization
```
