Default embedding dimension will be provided, incase if we need to change dimension then we can do it

```python
from langchain_openai import OpenAIEmbeddings
embeddings_1024 = OpenAIEmbeddings(model="text-embedding-3-large", dimensions=1024)
text = "This is a turorial on OPENAI embedding"
query_result = embeddings_1024.embed_query(text)
len(query_result)
```

#### Lets load text file and convert it into embeddings

```python
## Text Loader
from langchain_community.document_loaders import TextLoader

loader = TextLoader('/Users/hyashwanth/Desktop/Learning_Gen_AI/Generative-AI-Learnings/Lang_Chain/1-Langchain//Data_Transformer/datasets/speech.txt')
docs = loader.load()
docs
```

```python
## Text split

from langchain_text_splitters import RecursiveCharacterTextSplitter

text_splitter = RecursiveCharacterTextSplitter(chunk_size=500, chunk_overlap=50)
final_documents = text_splitter.split_documents(docs)
final_documents
```

```python
## Creating Embedding and store it in vector DB
## Vector Embedding and Vector storeDB (Chroma DB)
from langchain_community.vectorstores import Chroma

db = Chroma.from_documents(final_documents, embeddings_1024)
db
```

```python
## Retrieve the results from query vectorstore db
query = "It will be all the easier for us to conduct ourselves as belligerents"
retrieved_results = db.similarity_search(query)
print(retrieved_results)
```
